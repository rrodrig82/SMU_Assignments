```python
import pandas as pd
```

```python
my_csv_dfs = []
for n in range (1,13):
    url = f"{n}-citibike-tripdata.csv"
    my_csv_dfs.append(pd.read_csv(url))
```

```python
#combine all monthly csvs into one dataframe
main_df = pd.concat(my_csv_dfs)
```

```python
# check dataframe info
main_df.info()
```

```python
# take random sample
sub_df = main_df.sample(frac=0.01, replace=False, random_state=7)
len(sub_df)
```

```python
# reset dataframe index to month 
sub_df = sub_df.reset_index(drop=True)
sub_df["Month"] = [x.split("-")[1] for x in sub_df.starttime]
```

```python
# group dataframe by rides per month
sub_df.groupby('Month').count()
```

```python
# export dataframe to csv file 
sub_df.to_csv(r'sample_bike1.csv',index=False)
```
